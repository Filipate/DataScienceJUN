```python
# delete the duplicates in item_name and quantity

chipo.dropped = chipo.copy()
chipo_dropped= chipo.dropped.drop_duplicates(["item_name", "quantity"])
chipo_dropped
```

```python
## select only the products with quantity equals to 1
chipo_dropped = chipo_dropped[chipo_dropped.quantity == 1]
chipo_dropped
```

```python
# select only the item_name and item_price columns

chipo_dropped = chipo_dropped.loc[:, ["item_name", "item_price"]]
chipo_dropped
```

```python

# sort the values from the most to less expensive

chipo_dropped=chipo_dropped.sort_values("item_price", ascending=False)
chipo_dropped
```
